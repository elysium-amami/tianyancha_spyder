```python
def get_company(keys):
    '''
    keys为公司名称的字符串列表
    '''
    for id, key in enumerate(keys):
        url='https://www.tianyancha.com/search?key='+key
        driver.get(url)
        time.sleep(4)

        select=scrapy.Selector(text=driver.page_source)
        # 搜索页xpath路径
        company_result_list = '//*[@id="search_query"]//div[@class="result-list sv-search-container"]'
        # 选择第一个公司
        url=select.xpath(company_result_list + '/div[1]/div/div[3]/div[1]/a/@href').extract_first()
        driver.get(url)
        time.sleep(4)

        select=scrapy.Selector(text=driver.page_source)

        company={}
        company['基本信息']=get_base_info.get_base(select)
        company['工商信息']=get_business_info.get_gongshang(select)
        company['股东信息']=get_shareholders.get_gudong(select=select)
        company['主要成员']=get_main_members.get_zhuyao(select)
        company['企业年报']=get_anaul_report.get_report(select)
        company['经营风险']=get_operating_risk.get_operating_risk(select)
        company['司法风险']=get_judicial_risk.get_judicial_risk(select)
        company['知识产权']=get_intellectual_property.get_intellectual_property(select)

        for key, value in company.items():
            if type(value) == dict:
                for k, v in value.items():
                    col_dict[k].append(v)
            else:
                col_dict[key].append(value)
                
#        print('Done Company: ', id+1)
#         if id+1 % 100 == 0:
#             pd.DataFrame(col_dict).to_csv(f'company_{id+1}.csv', index=False)
    return  pd.DataFrame(col_dict)        
```

```python
keys = list(pd.read_csv('1000_want.csv')['名称'])
keys_lists = [keys[i: i + 50] for i in range(0, 1000, 50)]
# company_1000 = get_company(keys)
```

```python
for i, keys_list in enumerate(keys_lists):
    print(f'开始爬取第{i+1}组公司')
    group = get_company(keys_list)
    group.to_csv(f'company_group{i+1}.csv', index=False)
    print(f'完成爬取第{i+1}组公司')
    time.sleep(60)
    
```
